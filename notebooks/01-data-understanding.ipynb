## 4. Missing Value Assessment

Missing values are especially important in process data because they may arise from:

- sensor dropouts
- maintenance events
- logging interruptions
- target sparsity (for example, values only recorded near fault events)

Understanding where missingness occurs helps guide later preprocessing choices.

````python
missing_df = (
    df.isna()
      .sum()
      .reset_index()
      .rename(columns={"index": "column", 0: "missing_count"})
)

missing_df["missing_pct"] = 100 * missing_df["missing_count"] / len(df)
missing_df = missing_df.sort_values("missing_pct", ascending=False)

missing_df
````

````python
plt.figure(figsize=(10, 6))
sns.barplot(data=missing_df[missing_df["missing_count"] > 0], x="missing_pct", y="column")
plt.title("Missing Value Percentage by Column")
plt.xlabel("Missing Percentage (%)")
plt.ylabel("Column")
plt.tight_layout()
plt.show()
````

## Missing Value Interpretation

The missing value analysis reveals two distinct patterns in the dataset:

### 1. Extreme Missingness in `time_to_fault_min`
The variable `time_to_fault_min` has approximately **96.6% missing values**, which is significantly higher than all other variables.

This is not necessarily a data quality issue. In industrial process datasets, such variables are often only recorded:
- close to fault events
- during failure windows
- or when maintenance conditions are triggered

**Implication:**
- This variable should **not be used directly in the fault classification model**, as it may introduce data leakage.
- Instead, it is better suited for a **separate regression or predictive maintenance task**, where only rows with valid values are used.

---

### 2. Moderate Missingness in Sensor Variables (~6%)
Most process variables (e.g., temperature, pressure, flow rates, vibration, efficiency) show a consistent missing rate of approximately **6%**.

**Implication:**
- These values are manageable and can be handled during preprocessing.
- Forward-filling (time-aware imputation) or median-based imputation may be appropriate depending on context.
- Since missingness is relatively uniform across variables, it is unlikely to bias any single feature disproportionately.

## 6. Target Variable Review: `fault_type`

For classification, the target variable is central. We need to know:

- how many unique fault classes exist
- whether normal operation dominates
- whether the dataset is imbalanced

Class imbalance is expected in industrial fault datasets because systems typically spend most of their time in healthy operation.

````python
fault_counts = df["fault_type"].value_counts(dropna=False).reset_index()
fault_counts.columns = ["fault_type", "count"]
fault_counts["percentage"] = 100 * fault_counts["count"] / len(df)

fault_counts
````

````python
plt.figure(figsize=(10, 6))
sns.barplot(data=fault_counts, x="count", y="fault_type")
plt.title("Fault Type Distribution")
plt.xlabel("Count")
plt.ylabel("Fault Type")
plt.tight_layout()
plt.show()
````

The distribution of `fault_type` reveals a **highly imbalanced dataset**, which is a common characteristic of industrial process monitoring systems.

---

### 1. Dominance of Normal Operation

The majority class (`fault_type = 0`) accounts for approximately **96.4% of all observations**.

**Interpretation:**
- The system operates in a **healthy state most of the time**
- Faults are **rare events**, which is expected in well-maintained industrial processes
- This confirms that the dataset reflects **realistic operational conditions**

---

### 2. Minority Fault Classes

All fault classes (1, 2, 3, 4) individually represent **less than 2%** of the dataset:

- Fault 1: ~1.18%
- Fault 4: ~0.96%
- Fault 2: ~0.88%

**Interpretation:**

### 3. Severe Class Imbalance Problem

This is a textbook example of **extreme class imbalance**.

If not handled correctly:
